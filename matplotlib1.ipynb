```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x= np.arange(0,10)
y = np.random.randint(0,50,size=10)
y1= np.random.randint(0,23,size=10)

plt.plot(x,y,label ='y')
plt.plot(x,y1,label='Y1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

# scatter plot --- we use scatter plot when x values don't have any particular sequence.

```python
x2= [1,10,2,32,43,22,33,43,12,32,22,65,75,34,23,43,13,43]
y2 = np.random.randint(1,50,size=18)
fig1 = plt.figure()
plt.plot(x2,y2)
plt.show()

fig2 = plt.figure()
plt.scatter(x2,y2)
plt.show()
```

```python
scatter_data = pd.DataFrame({
    'x1' : np.random.randint(1,200,size=1000),
    'sales' : np.random.randint(1,200,size=1000),
    's1' : np.random.randint(30,130,size=1000),
    'a1' : np.random.random(size=1000),
    'zones':np.random.randint(0,3,size=1000)
              })
```

```python
plt.scatter(x=scatter_data['x1'],
            y=scatter_data['sales'],
            s=scatter_data['sales'],
            c=scatter_data['zones'],
            alpha = 0.6
            )
```

```python
mt = pd.read_csv('mtcars.csv')
plt.scatter(mt['mpg'], mt['hp'], c=mt['am'],alpha=1)
```

```python
plt.pie([23,34,78,19],
        labels=list('NSEW'),
        colors=['teal','skyblue','blue','#3f2f75'],
        autopct="%1.0f%%",
        explode=[0,0,0,0.2],
        shadow=True
        )
plt.show()
```
